```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

# Load your dataset
df = pd.read_csv('Auto Sales data.csv')

# Selecting relevant numerical columns
analysis_df = df[['QUANTITYORDERED', 'MSRP', 'SALES']]

# Measuring strength
corr_matrix = analysis_df.corr()
print("--- STEP 2: CORRELATION COEFFICIENTS ---")
print(corr_matrix['SALES'].round(4))
```

```python
# Feature and Target selection
X = analysis_df[['QUANTITYORDERED']] 
Y = analysis_df['SALES']            

model = LinearRegression()
model.fit(X, Y)

m = model.coef_[0]
c = model.intercept_
r_sq = model.score(X, Y)

print("\n--- STEP 3: REGRESSION PARAMETERS ---")
print(f"Slope (m): {m:.4f}")         
print(f"Intercept (c): {c:.4f}")     
# FIX: Using f-string formatting instead of .round()
print(f"R-Squared (Accuracy): {r_sq:.4f}")
```
