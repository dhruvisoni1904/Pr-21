```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

# Load your dataset
df = pd.read_csv('Auto Sales data.csv')

# Selecting relevant numerical columns
analysis_df = df[['QUANTITYORDERED', 'MSRP', 'SALES']]

# Measuring strength
corr_matrix = analysis_df.corr()
print("--- STEP 2: CORRELATION COEFFICIENTS ---")
print(corr_matrix['SALES'].round(4))
```

```python
# Feature and Target selection
X = analysis_df[['QUANTITYORDERED']] 
Y = analysis_df['SALES']            

model = LinearRegression()
model.fit(X, Y)

m = model.coef_[0]
c = model.intercept_
r_sq = model.score(X, Y)

print("\n--- STEP 3: REGRESSION PARAMETERS ---")
print(f"Slope (m): {m:.4f}")         
print(f"Intercept (c): {c:.4f}")     
# FIX: Using f-string formatting instead of .round()
print(f"R-Squared (Accuracy): {r_sq:.4f}")
```

```python
# --- STEP 4: PREDICTION (FIXED) ---
# Instead of a NumPy array, we create a small DataFrame with the correct column name
new_data = pd.DataFrame({'QUANTITYORDERED': [50]})

# Predict sales
prediction = model.predict(new_data)

print(f"--- STEP 4: PREDICTION ---")
print(f"Predicted Sales for 50 Units: ${prediction[0]:.2f}")
```

```python
plt.figure(figsize=(10, 5)) 

# Plot 1: Correlation Heatmap 
plt.subplot(1, 2, 1) 
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f") 
plt.title("Correlation Map") 

# Plot 2: Regression Line 
plt.subplot(1, 2, 2) 
sns.regplot(x='QUANTITYORDERED', y='SALES', data=analysis_df, scatter_kws={'alpha':0.3}, line_kws={"color": "red"}) 
plt.title("Regression: Quantity vs Sales") 

plt.tight_layout() 
plt.show()
```
